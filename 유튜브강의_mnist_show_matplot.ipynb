# mnist data Image 보여주기_matplotlib

```python
import sys, os
# sys.path.append(os.pardir) : 부모 디렉터리의 파일을 가져올 수 있도록 설정
# sys.path : 파일을 불러올 때 sys.path에 등록되어있는 경로만 뒤져서 파일을 찾아옴
# append : 리스트를 추가할 때 사용하는 매서드
# os.pardir : 상위 폴더
# cf) os.getcwd() : 현재 작업하고 있는 디렉토리 위치를 보여줌(= pwd)
sys.path.append("./dataset") # 이때, dataset 폴더는 실행하는 py 파일의 경로와 일치해야 한다.
import numpy as np
from dataset.mnist import load_mnist

# 이미지 처리를 해주는 라이브러리
import matplotlib.pyplot as plt
```

```python
(x_train, t_train), (x_test, t_test) = load_mnist(flatten=False, normalize=False)
```

```python
# figure : 도화지
plt.figure()
# imshow : 행렬을 사람이 볼 수 있는 이미지로
plt.imshow(x_train[0][0])
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(10,10))
```

```python
# xticks : 눈금
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(x_train[i][0], cmap=plt.cm.binary)
    plt.xlabel(t_train[i])
```
